```python
X = df.drop(['Pass/Fail'], axis = 1)
y = df['Pass/Fail']
```

```python
# Do train-test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8, random_state = 101)
```

```python
# Create function for model_accuracy
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score

model = RandomForestClassifier(n_estimators = 100, n_jobs = -1, random_state = 101)

def model_accuracy(X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred1 = model.predict(X_train)
    return ('Accuracy Score (Train data)', accuracy_score(y_train, y_pred1)*100,
            'Accuracy Score (Test data): ', accuracy_score(y_test, y_pred)*100,
            'ROC AUC score: ', roc_auc_score(y_test, y_pred))
```

```python
# Create Function for Classification Performance

from sklearn.metrics import classification_report, confusion_matrix

# Here is for confusion matrix
def conf_matrix(X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    matrix = sn.heatmap(confusion_matrix(y_test,y_pred), cmap = 'Blues', annot=True)
    return matrix

# Here is for classification report
def class_matrix(X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print (classification_report(y_test,y_pred))
```

```python
# check for model accuracy

model_accuracy(X_train, X_test, y_train, y_test)
```

```python
conf_matrix(X_train, X_test, y_train, y_test)
```

```python
class_matrix(X_train, X_test, y_train, y_test)
```

```python
# Check for any multicollinearity from correlation matrix
corr_matrix = df1.corr()
# corr_matrix
```

```python
# Plot the heatmap
sn.heatmap(corr_matrix)
```

```python
# Multicollinearity are hardly seen through this heatmap.
```

```python
# In the correlation heatmap, data description is written twice because categories are repeated on both axis.
# A correlation map that present data description without repetition is called Triangle Correlation heatmap.
```

```python
# Create Triangle correlation heatmap (can be upper trianggle/lower triangle)
# Now, upper triangle correlation heatmap is used

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
# upper
```

```python
# Plot the upper triangle heatmap
sn.heatmap(upper)
```

```python
# Multicollinearity happens when the correlation of 2 variables > 0.9 or < -0.9

# Find multicollinearity (corr>0.9)
pos_corr = [column for column in upper.columns if any(upper[column]>0.9)]
print(pos_corr) # Return the list of column indexes with correlation > 0.9

# Find multicollinearity (corr<-0.9)
neg_corr = [column for column in upper.columns if any(upper[column]<-0.9)]
print(neg_corr) # Return the list of column indexs with correlation <-0.9
```

```python
# Sum all columns with multicollinearity
tot_corr = pos_corr + neg_corr
len(tot_corr) # There are 226 columns with multicollinearity
```

```python
# Drop Columns with multicollinearity
df2 = df1.drop(columns = tot_corr, axis = 1)
df2.shape
```

```python
# Construct the model

X2 = df2.drop(['Pass/Fail'],axis = 1)
y2 = df2['Pass/Fail']

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2,y2,train_size = 0.8, random_state = 101)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
```

### RFE Without SMOTE() techniques

```python
# Perform Stratification Random Sampling

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2,y2,train_size = 0.8, 
                                                        random_state = 101, stratify = y)
```

```python
X_train2.shape, X_test2.shape, y_train2.shape, y_test2.shape
```

```python
# To use RFE, we have to provide the base model and let RFE help to do feature selection.
# Now, we use Random Forest as the base model.
```

```python
# Construct Random Forest model

sfm = SelectFromModel(RandomForestClassifier(n_estimators = 100, n_jobs=-1, random_state = True))

# SelectFromModel - select features which importance are greater than the mean importance of features.

sfm.fit(X_train2, y_train2)
# sfm.get_support()
#get.support() returns boolean values which reveal the important features.

sfm.estimator_.feature_importances_ # to see the feature importance scores
```

```python
# Show the table of significant features

feature_selected = pd.DataFrame(X_train2.columns[sfm.get_support()])
feature_selected # this is list of the significant columns
```

```python
len(feature_selected) # There is a total of 100 significant variables.
```

```python
# SFM for train data
df3_train = sfm.transform(X_train2)
df3_train = pd.DataFrame(df3_train)
df3_train.shape
```

```python
# SFM for test data
df3_test = sfm.transform(X_test2)
df3_test = pd.DataFrame(df3_test)
df3_test.shape
```

```python
# Check for model accuracy

model_accuracy(df3_train, df3_test, y_train2, y_test2)
```

```python
class_matrix(df3_train, df3_test, y_train2, y_test2)
```

```python
conf_matrix(df3_train, df3_test, y_train2, y_test2)
```

### RFE With SMOTE() Techniques

```python
# Since we know that the data is imbalance, we can perform SMOTE() technique to balance the data.
# SMOTE (Synthetic Minority Oversampling Techniques)
```

```python
# SMOTE techniques is only applied in train data
from imblearn.over_sampling import SMOTE
```

```python
osmote = SMOTE()
Xs_train, ys_train = osmote.fit_resample(df3_train, y_train2)
```

### VIF without SMOTE() Techniques
* Try to Perform VIF, instead of RFE after df2

```python
#Do Standardization first
features = df2.drop(['Pass/Fail'], axis = 1)
target = df['Pass/Fail']
col = list(features.columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_feature = scaler.fit_transform(features)
scaled_feature = pd.DataFrame(scaled_feature, columns=col)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['variables']=scaled_feature.columns
vif['VIF'] = [variance_inflation_factor(scaled_feature.values,i) for i in range(scaled_feature.shape[1])]
```

```python
# The significant columns are column with VIF score not larger than 10

high_vif = vif[vif['VIF']<10].sort_values('VIF')
high_vif
```

```python
df3_vif = scaled_feature[high_vif['variables']]
df3_vif
```

```python
X_train3, X_test3, y_train3, y_test3=train_test_split(df3_vif, target, train_size = 0.8, random_state = 101)
```

```python
model_accuracy(X_train3, X_test3, y_train3, y_test3)
```

```python
class_matrix(X_train3, X_test3, y_train3, y_test3)
```

```python
conf_matrix(X_train3, X_test3, y_train3, y_test3)
```

### Findings:
* The significant columns left after RFE and VIF are 100 and 189, respectively.Dimensionality reduction by VIF creates slightly better recall, precision and f-1 score values than that by RFE. It may be caused by higher number of column in VIF. So, dimensionality reduction by RFE is preferred. slightly lower classification metrics score, but lower significant column needed.
* If we compare the classification by RFE with and without oversampling method, it is shown that the classification metrics are lower with oversampling method. So, oversampling method is not needed.
